```python
import gym

def run_cartpole():
    # Create the CartPole environment with render_mode specified
    env = gym.make('CartPole-v1', render_mode='human')

    # Run the environment for a few episodes
    for episode in range(100):  # Adjust the number of episodes as needed
        # Reset the environment to its initial state for each episode
        observation = env.reset()

        # Run the environment for a few time steps
        for _ in range(1000):
            # Render the environment
            env.render()

            # Take a random action (replace this with your reinforcement learning logic)
            action = env.action_space.sample()

            # Perform the action and get the new state
            step_result = env.step(action)

            # Handle the returned value appropriately
            if isinstance(step_result, tuple) and len(step_result) >= 3:
                observation, reward, done = step_result[:3]
            else:
                raise ValueError("Unexpected return format from env.step")

            # Check if the episode is done (the pole has fallen or the time limit is reached)
            if done:
                print(f"Episode {episode + 1} finished!")
                break

            # Explicitly reset the environment if terminated is True
            if done:
                env.reset()

    # Do not close the environment after the loop

if __name__ == "__main__":
    run_cartpole()
```
